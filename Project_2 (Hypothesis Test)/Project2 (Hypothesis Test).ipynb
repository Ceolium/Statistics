# [Hypothesis test results and findings]

## Statistical Tests

### Q1. The average bending strength of the wooden beams produced by the company is below 50N/mm2.
#### a) Z-test on large sample data (douglas data 50.csv) (n>30)<br>
> As the question says the mean is lowers than the 50, the null hypothesis should be H0 : μ >= 50 and alternative hypothesis should be Ha : μ < 50. The size of sample is bigger than 30, so we can apply Z-test to this data.

- **The significant level: α = 0.05**
- **H0 : μ >= 50 <br>**
-  **Ha : μ < 50 **

```python
mu0 = 50
alpha = 0.05

zstat = (xbar50 - mu0) / (sig50/np.sqrt(N50))
print 'z-statistic:', round(zstat,3)
pval = (1 - norm.cdf(abs(zstat)))
print 'p-value:', round(pval,3)

x = np.linspace(0,100,100)
y = norm.pdf(x, loc=mu0, scale=sig50)
plt.plot(x,y)

perc05 = norm.ppf(alpha, loc=mu0, scale=sig50)
plt.axvline(perc05, color='r')
plt.axvline(xbar50, color='k', lw=2)

if pval < alpha:
    plt.title('Enough evidence to reject H0')
else:
    plt.title('Not enough evidence to reject H0')
    
plt.show()
```

#### Z-test Interpretation

> 
- z-statistic: **-0.481** / p-value: **0.315**
- Null hypothesis: μ >= 50, this should be **one-tailed test.**
- As **p-value < alpha**, we **cannot reject the null hypothesis.**
- This means that the given evidence cannot defy the company's report

#### b) t-test on small sample data (douglas data 20.csv) (n < 30)
> As the question says the mean is lowers than the 50, the null hypothesis should be H0 : μ >= 50 and alternative hypothesis should be Ha : μ < 50. The size of sample is smaller than 30, so we should apply t-test to this data.


- **The significant level: α = 0.05**
- **H0 : μ >= 50** <br>
- **Ha : μ < 50**

```python
mu0 = 50
alpha = 0.05
dof = N50 - 1

tstat = (xbar50 - mu0) / (sig50/np.sqrt(N50))
print 't-statistic:', round(tstat,3)

pval = t.sf(np.abs(tstat), dof)
print 'p-value:', round(pval,3)

x = np.linspace(0,100,100)
y = t.pdf(x, dof, loc=mu0, scale=sig50)
plt.plot(x,y)


perc05 = t.ppf(alpha, dof, loc=mu0, scale=sig50)
calAlpha = t.cdf(perc05, dof, loc=mu0, scale=sig50)
print 'This is the verificaton of alpha value:', calAlpha

plt.axvline(perc05, color='r')
plt.axvline(xbar50, color='k', lw=2)


if pval < alpha:
    plt.title('Enough evidence to reject H0')
else:
    plt.title('Not enough evidence to reject H0')
    
plt.show()
```

#### t-test Interpretation

> 
- t-statistic: **-0.481** / p-value: **0.316** <br>: We can observe that the t-statistic quite similar to z-statistic. So does the p-value.
- Null hypothesis: μ >= 50, this should be **one-tailed test.** 
- As **p-value < alpha**, we **cannot reject the null hypothesis.**
- This means that the given evidence cannot defy the company's report

### Q2. There is no difference between the average bending strength of wooden beams produced in Eastern Canada and that of wooden beams produced in Western Canada.

#### z-test on sample datasets from two populations<br>
> As the question says the there is no difference in average, the null hypothesis should be H0 : μ(west) = μ(east) and alternative hypothesis should be Ha : μ(west) ≠ μ(east). The size of both samples are larger than 30, so we can apply Z-test to these two data(douglas eastern data 42.csv & douglas weatern data 30.csv)

- **The significant level: α = 0.05**
- **H0 : μ(west) = μ(east) **<br>
- **Ha : μ(west) ≠ μ(east)**

```python
alpha = 0.05

zstat = (xbar42 - xbar30) / np.sqrt((sig42)*(sig42)/N42 + (sig30)*(sig30)/N30)
print('z-statistic: ', round(zstat,3))
pval = 2*(1 - norm.cdf(abs(zstat)))
print('p-value: ', round(pval,4))


x = np.linspace(0,100,100)
y = norm.pdf(x, loc=xbar30, scale=np.sqrt((sig42)*(sig42) + (sig30)*(sig30)))

plt.plot(x,y)

perc05 = norm.ppf(alpha, loc=xbar30, scale=np.sqrt((sig42)*(sig42) + (sig30)*(sig30)))


plt.axvline(perc05, color='r')
plt.axvline(xbar42, color='k', lw=2)

if pval < alpha:
    plt.title('Enough evidence to reject H0')
else:
    plt.title('Not enough evidence to reject H0')

plt.show()
    
```

```python
x = np.linspace(-3, 3, 100)
norm_pdf = lambda x: (1/np.sqrt(2 * np.pi)) * np.exp(-x * x / 2)
y = norm_pdf(x)

fig, ax = plt.subplots(1, 1, sharex=True)
ax.plot(x, y)
ax.fill_between(x, 0, y, where = x > zstat)
ax.fill_between(x, 0, y, where = x < -zstat)
plt.title('Rejection regions for a two-tailed hypothesis test at 95% confidence')
plt.xlabel('x')
plt.ylabel('p(x)');
plt.show()
```

#### Two samples Z-test Interpretation

> 
- z-statistic: **2.209** / p-value: **0.0272** 
- Null hypothesis: H0 : μ(west) = μ(east), this should be **two-tailed test.** 
- As **p-value > alpha**, we **can reject the null hypothesis.**
- This means that the given evidence **contradicts** the company's report saying that there is no difference between the average bending strength of wooden beams produced in Eastern Canada and that of wooden beams produced in Western Canada.

### Q3. There is no difference between the variance of bending strength of wooden beams produced in Estern Canada and that of wooden beams produced in Western Canada.
#### F-test on sample datasets from two populations<br>
> As the question says the there is no difference in variance, the null hypothesis should be H0 : Var(west) = Var(east) and alternative hypothesis should be Ha : Var(west) ≠ Var(east). To compare the variance, we should apply F-test to these two data(douglas eastern data 42.csv & douglas weatern data 30.csv)

- **The significant level: α = 0.05**
- **H0 : Var(west) = Var(east)** <br>
- **Ha : Var(west) ≠ Var(east) **

```python
print '[Sample 42]'
print 'sample mean:', round(xbar42,3)
print 'standard deviation:', round(sig42,3)
print '[Smaple 30]'
print 'sample mean:', round(xbar30,3)
print 'standard deviation:', round(sig30,3)
print '\n'

alpha = 0.05
dof1 = N42 - 1
dof2 = N30 - 1
Fstat = np.var(pop_42.bstrength) / np.var(pop_30.bstrength)
Two_sided_pval = 2*min(f.cdf(Fstat, dof1, dof2), 1-f.cdf(Fstat, dof1, dof2))

print "Degree of Freedom(1): %i" % (dof1)
print "Degree of Freedom(2): %i" % (dof2)
print "F-statistic : %.3f" % (round(Fstat,3))
print "p-value : %.3f" % (round(Two_sided_pval,3))
if Two_sided_pval < alpha:
    print ('enough evidence to reject H0')
else:
    print ('Not enough evidence to reject H')
```

#### F-test Interpretation

> 
- F-statistic: **0.985** / p-value: **0.948** 
- Null hypothesis: H0 : Var(west) = Var(east), so this should be **two-tailed test.** 
- As **p-value < alpha**, we **cannot reject the null hypothesis.**
- This means that the given evidence cnoot defy the company's report saying that there is no difference between the variance of bending strength in Eastern Canada and Western Canada.
